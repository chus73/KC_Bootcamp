```python
import scrapy
import json
from datetime import timedelta, datetime

start, end = datetime(2001, 1, 1), datetime(2019, 7, 9)
root = 'https://www.dailyfx.com/calendar?previous=true&currentweek=currentweek&week='
week_start = start - timedelta(days=start.isoweekday() % 7)
week_end = end - timedelta(days=end.isoweekday() % 7)
    
class BlogSpider(scrapy.Spider):
    name = 'blogspider'
    dates = [week_start + 7 * timedelta(days=i) for i in range(int(((week_end - week_start).days + 1) / 7))]
    start_urls = []
    for element in dates:
        start_urls.append(root + str(element.year) + '/' + str(element.month).zfill(2) + str(element.day).zfill(2))

    def parse(self, response):
        for row in response.css('#daily-cal1 tr.event'):
            fecha = row.css('td:nth-child(2) ::text').extract()
            evento = row.css('td:nth-child(4) ::text')[-1].extract()
            importancia = row.css('td:nth-child(5) span:first-child ::text').extract_first()
            actual = row.css('td:nth-child(6) ::text').extract_first()
            prevision = row.css('td:nth-child(7) ::text').extract_first()
            previo = row.css('td:nth-child(8) ::text').extract_first()
            prevision_previo = row.css('td:nth-child(9) ::text').extract_first()
            # yield{'sale:':row.css('div.eco-table-date ::text').extract_first().strip}
            print(f'{fecha}|{evento}|{importancia}|{actual}|{prevision}|{previo}|{prevision_previo}\n', file=filep)

print("Definici√≥n parse lista")

from scrapy.crawler import CrawlerProcess

process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'
})

filep = open('data.csv', 'w')
process.crawl(BlogSpider)
process.start()
filep.close()
print("Crawler ejecutado")
```
